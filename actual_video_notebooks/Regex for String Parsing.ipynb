```python
import re
```

```python
word = r'\w+'
```

```python
sentence = "I love using Regex."
```

```python
re.findall(word, sentence)
```

```python
search_result = re.search(word, sentence)
```

```python
search_result
```

```python
search_result.group()
```

```python
match_result = re.match(word, sentence)
```

```python
match_result
```

```python
match_result.group()
```

```python
capitalized_word = r'[A-Z]\w+'
```

```python
search_result = re.search(capitalized_word, sentence)
```

```python
search_result
```

```python
search_result.group()
```

```python
match_result = re.match(capitalized_word, sentence)
```

```python
match_result
```

```python
match_result.group()
```

```python
sentence_with_digits = 'Even though I live 60,235 meters from the airport, I still here at least 4 planes.'
```

```python
sentence_with_digits
```

```python
numbers = r'\d+'
```

```python
re.findall(numbers, sentence_with_digits)
```

```python
thousands_number = r'(\d+,\d+|\d+)'
```

```python
re.findall(thousands_number, sentence_with_digits)
```

```python
city_state = r'(?P<city>[\w\s]+), (?P<state>[A-Z]{2})'
```

```python
my_address = 'My House, 123 Main Street, Los Angeles, CA 90013'
```

```python
for city_match in re.finditer(city_state, my_address):
    print(city_match.group('city'))
```
